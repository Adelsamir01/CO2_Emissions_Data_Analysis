```python
# importing liberaries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing the datasets
countries = pd.read_csv('data/countries.tsv', sep='\t')
missions = pd.read_csv('data/missions.tsv', sep='\t')
places = pd.read_csv('data/places.tsv', sep='\t')
users = pd.read_csv('data/users.tsv', sep='\t')
```

### Assessing the Data

```python
countries.head()
```

```python
missions.head()
```

```python
places.head()
```

```python
users.head()
```

```python
# renaming some columns to match and merge all datasets into one 
users = users.rename(columns={'#user_id': 'user_id','name':'user_name'})
places = places.rename(columns={'#place_id' :'place_id', 'country':'#alpha2'})
```

```python
# mergeing all datasets into one main dataset 
main = pd.merge(places, countries, how='inner',on=['#alpha2'])
main = pd.merge(main, missions, how='inner', on=['place_id'])
main = pd.merge(main, users, how='inner', on=['user_id'])
```

```python
# Spliting date into Year, Month, Day for analysis
main['date'] = pd.to_datetime(main['date'], errors='coerce')
main['day'] = main['date'].dt.day
main['month'] = main['date'].dt.month
main['year'] = main['date'].dt.year
```

```python
# Dropping Unessessary Columns
main = main.drop(columns=['place_id','#alpha2','#mission_id','user_id','date'])
```

```python
# saving main into csv file 
main.to_csv('combined_dataset.csv')
```

## Comparing Subsets

**We know from General insigts that France is the highest in carbon emissions, but what are the highest hauses in France?**

```python
main[main['country']=='France'].groupby(["house"]).co2.sum().reset_index().sort_values(['co2'], ascending=False).head(10)
```

**Now, we need to define the relationship between the duration of trip and the CO2 emissions**

```python
main.co2.corr(main.duration)
```

the correlation coefficient between duration of the trip and the CO2 is small, thus we conclude that co2 emissions is not dependent or linerarly related to the trip duration

**We know from the global insights that France comes first in co2 emissions and then comes the United States. But does it have something to do with the year? Below is the highest 3 years of emissions only in France and only in United States (Eachone Separetly)**

```python
# top 3 years in co2 emissions in France
main[main['country']=='France'].groupby(["year"]).co2.sum().reset_index().sort_values(['co2'], ascending=False).head(3)
```

```python
# top 3 years in co2 emissions in United States
main[main['country']=='United States'].groupby(["year"]).co2.sum().reset_index().sort_values(['co2'], ascending=False).head(3)
```

Now, it's pretty clear that 2015 and 2016 made it to the top 3 years in both countries, which points out that more activity in these two years was clearly an indicator in having more co2 emissions, in two different countries, not depending on the place.

## Visualization 

**Now, we need to make a correlation map between the Trip Duration, Distance, and Co2 emissions to be able to detect a clear linear relationship that could spot the light on the main reason of co2 increased emissions.**

```python
# creating correlation map to investigate the relation between duration, distance, co2
import seaborn as sns

f,ax = plt.subplots(figsize=(7, 7))
sns.heatmap(main[['co2', 'distance', 'duration']].corr(), annot=True, linewidths=.5, fmt= '.1f')
plt.title('Correlation');
```

Now, it is clear that it is not the distance nor the duration imply the increased emissions of the co2 with a low correlation coofficient of 0.1 and 0.2

**However, to make further analysis to clarify the relationship between co2 emissions and time**

```python
# destribution of co2 emissions over time (line chart)
co2_year = main.groupby(["year"]).co2.sum().reset_index().sort_values(['year'], ascending=False)
plt.rcParams['figure.figsize']=(15,9)


submain7 = main.groupby(["year"]).co2.sum().reset_index().sort_values(['year'], ascending=False)
    
plt.bar(submain7.year, submain7.co2 , data = submain7, color=['black', 'red', 'green', 'blue', 'cyan','yellow','gold'])
plt.xlim((2013,2020))
plt.ylim((0,100))
plt.xlabel('Year')
plt.ylabel('CO2 Emisions')
plt.title('Co2 emissions over years');
  
plt.show()
```

It's clear that at 2014 co2 emissions had a huge spike and continuing to increase through 2015, but since then it has been decreasing. That could be due to shifting into alternative energy sources and country regulations.

```python
def filter_country(search_term):
    '''
    Filters the medals_df DataFrame to only contain rows whose Sport or Discipline
    columns contain the given search_term.
    '''
    return main[main["country"].str.contains(search_term, case=False) \
                     | main["region"].str.contains(search_term, case=False)\
                     | main["house"].str.contains(search_term, case=False)]
```

```python
def plot_countries(main):
    co2_year = main.groupby(["year"]).co2.sum().reset_index().sort_values(['year'], ascending=False)
    plt.rcParams['figure.figsize']=(15,4)

    co2_year.plot(kind = 'line', x = 'year', y = 'co2', alpha = 0.5, color = 'blue')
    plt.xlim((2013,2020))
    plt.ylim((0,100))
    plt.xlabel('Year')
    plt.ylabel('CO2 Emisions')
    plt.title('Co2 emissions over years');
    
    
```

```python
# import display to show DataFrame content
from IPython.display import display
# import the widgets interact function to automatically create interactive UI
from ipywidgets import interact

# decorate the search_and_plot_medals function using interact
# to create text field UI for search term
@interact(search='')
def search_and_plot_countries(search):
    '''
    Display medals by country when given a search term.
    '''
    filtered_country = filter_country(search)
    if len(filtered_country) == 0:
        print('No hits')
        return
    
    # show plot 
    plot_countries(filtered_country)
```

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

def f(Year): 
    
    submain1 = main[['continent','year','co2']]
    submain2 = submain1[submain1['year']==Year].groupby(["continent"]).co2.sum().reset_index()
    
    plt.bar(submain2.continent, submain2.co2 , data = submain2, color=['black', 'red', 'green', 'blue', 'cyan'])
    plt.xlabel("Continent")
    plt.ylabel("Co2 Emissions")
    plt.title("Breakdown of the co2 emissions for each contienant by year")
    plt.ylim(0, 90)
    plt.rcParams['figure.figsize']=(10,4)
    plt.show()

interactive_plot = interactive(f, Year=(2013,2020))
output = interactive_plot.children[-1]

interactive_plot
```

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

def f(Day): 
    
    submain3 = main[['country','day','co2']]
    submain4 = submain3[submain3['day']==Day].groupby(["country"]).co2.sum().reset_index().sort_values(['co2'], 
                                                                                                       ascending=False).head(5)
    
    
    
    fig, ax = plt.subplots(figsize=(15, 10), subplot_kw=dict(aspect="equal"))

    data = submain4.co2
    ingredients = submain4.country


    def func(pct, allvals):
        absolute = int(pct/100.*np.sum(allvals))
        return "{:.1f}%\n".format(pct, absolute)


    wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                      textprops=dict(color="w"))

    ax.legend(wedges, ingredients,
              title="Countries",
              loc="center left",
              bbox_to_anchor=(1, 0, 0.5, 1))

    plt.setp(autotexts, size=12, weight="bold")

    ax.set_title("Co2 destribution over top 5 countries based on days")

    plt.show()

    
    

interactive_plot = interactive(f, Day=(1,31))
output = interactive_plot.children[-1]

interactive_plot
```
